```python
import builtins
import base64
```

```python
def create_folder(pdf_path):
  base_name = os.path.splitext(os.path.basename(pdf_path))[0]
  output_dir = os.path.join(os.getcwd(), base_name)

  if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"Đã tạo thư mục: {output_dir}")
  else:
    print(f"Thư mục {output_dir} đã tồn tại")

  return output_dir
```

```python
def extract_pdf(pdf_path, output_dir, base_name):
  reader = PdfReader(pdf_path)
  total_pages = len(reader.pages)

  for page_index in range(total_pages):
    writer = PdfWriter()
    page = reader.pages[page_index]
    writer.add_page(page)

    output_pdf = os.path.join(output_dir, f"{base_name}_page_{page_index+1}.pdf")
    with open(output_pdf, "wb") as out_f:
        writer.write(out_f)
    print(f"Đã lưu trang {page_index + 1} tại: {output_pdf}")
```

```python
def sort_files(folder_path):
  if not os.path.exists(folder_path):
          raise FileNotFoundError(f"Thư mục {folder_path} không tồn tại")

  new_folder = []
  for file in os.listdir(folder_path):
      if file.endswith('.pdf'):
          new_folder.append(file)

  if not new_folder:
      raise ValueError(f"Không tìm thấy file PDF trong {folder_path}")

  try:
      sorted_files = builtins.sorted(new_folder, key=lambda x: int(x.split('_')[-1].split('.')[0]))
  except ValueError as e:
      print(f"Lỗi định dạng tên file: {e}")
      sorted_files = builtins.sorted(new_folder)  # Sắp xếp mặc định nếu lỗi
  print(sorted_files)
  return sorted_files
```

```python
def convert_img_to_text(sorted_files, base_name, folder_pdf):
  text = []
  for i in range(1, len(sorted_files)+1):
    obj = {}
    result = analyze_image(f"{folder_pdf}/img_convert/{base_name}_page_{i}.png")
    obj['page'] = i
    obj['text'] = result
    text.append(obj)
  return text
```

```python
from pdf2image import convert_from_path
def analyze_image(image_path):
    with open(image_path, "rb") as image_file:
        response = client.chat.completions.create(
            model="gpt-4o",
            messages=[
                {
                    "role": "user",
                    "content": [
                        {
                            "type": "text",
                            "text": "Trích xuất toàn bộ nội dung trong hình ảnh: chuyển đổi các công thức toán học thành định dạng LaTeX và giữ nguyên văn bản thông thường dưới dạng text. Trả về kết quả rõ ràng, phân biệt giữa văn bản và công thức."
                        },
                        {"type": "image_url", "image_url": {"url": f"data:image/png;base64,{base64.b64encode(image_file.read()).decode('utf-8')}"}}
                    ]
                }
            ],
            max_tokens=1000
        )
    return response.choices[0].message.content
```

```python
def visualize_graph(data):
    try:
        import networkx as nx
        from pyvis.network import Network
        from IPython.display import IFrame, display

        # Create a directed graph
        G = nx.DiGraph()

        # Process the data to populate the graph
        for record in data:
            # Extract node_a, node_b, and the relationship type
            node_a = record.get('node_a')
            node_b = record.get('node_b')
            relationship = record.get('r')[1] if 'r' in record and isinstance(record['r'], tuple) else None

            if node_a and node_b and relationship:
                # Add nodes and edge with the relationship type as a label
                G.add_node(node_a, label=node_a)
                G.add_node(node_b, label=node_b)
                G.add_edge(node_a, node_b, label=relationship)

        # Create a Pyvis Network object with CDN resources set to 'in_line'
        net = Network(notebook=True, directed=True)
        net.from_nx(G)

        # Add edge labels for better visualization
        for edge in G.edges(data=True):
            net.add_edge(edge[0], edge[1], title=edge[2].get('label', ''))

        # Save the graph to an HTML file
        net.show('graph.html')

        # Embed the graph in Google Colab using IFrame
        display(IFrame('graph.html', width='100%', height='100%'))

    except Exception as e:
        print(f"Error while visualizing the graph: {str(e)}")
        return None
```

```python
def add_relationships_to_neo4j(kg, relationships):
    """
    Add relationships extracted from the knowledge graph to the Neo4j database.
    """
    with kg._driver.session() as session:
        for subject, relation, obj in relationships:
            try:
                # Chuẩn hóa dữ liệu (chỉ loại bỏ dấu nháy nếu cần)
                subject = subject.replace("'", "").strip()
                relation = relation.replace("'", "").strip()
                obj = obj.replace("'", "").strip()

                # Chuyển relation thành định dạng hợp lệ cho Neo4j (thay khoảng trắng bằng gạch dưới)
                clean_relation = relation.replace(" ", "_").upper()
                if not clean_relation:
                    clean_relation = "RELATED_TO"  # Dự phòng nếu relation rỗng

                # Truy vấn Cypher để thêm quan hệ
                query = """
                MERGE (s:Entity {name: $subject})
                MERGE (o:Entity {name: $object})
                MERGE (s)-[r:%s]->(o)
                SET r.latex_name = $latex_name
                """ % clean_relation

                session.run(
                    query,
                    subject=subject,
                    object=obj,
                    latex_name=relation
                )
            except Exception as e:
                print(f"Lỗi khi thêm quan hệ ({subject}, {relation}, {obj}): {e}")
                continue

    print("Hoàn tất thêm quan hệ vào Neo4j.")
```

```python
#File kỹ năng sư phạm
data_simple = []

for pdf in pdf_path_kn:
  new_chunk = []

  base_name = os.path.splitext(os.path.basename(pdf))[0]
  folder_pdf = create_folder(pdf)
  print(folder_pdf)
  extract_pdf(pdf, folder_pdf, base_name)
  sort_filesssss = sort_files(folder_pdf)

  for filename in sort_filesssss:
      if filename.endswith(".pdf"):
          file_path = os.path.join(folder_pdf, filename)

          elements = partition_pdf(
                filename=file_path,
                strategy='hi_res',
                extract_images_in_pdf=False,  #
                languages=['vie'],
                chunking_strategy='by_title',
                partitioning_strategies='auto',
                multipage_sections=True,
                max_characters=2048,
                new_after_n_chars=2048,
                combine_text_under_n_chars=1024,
                max_chunk_overlap=256
              )

          for element in elements:
            data = {
              'page_content' : element.text,
              'metadata' : {
                  'source' : element.metadata.fields['file_directory'],
                  'page' : element.metadata.filename.split("_page_")[1].split(".")[0],
                  }
              }
            new_chunk.append(data)

  data_simple.append(new_chunk)
```

```python
def extract_entities(text):
    prompt = (
        f"Trích xuất các thực thể (entities) từ văn bản dưới đây.\n"
        f"Thực thể phải bằng tiếng Việt và theo định dạng:\n"
        f"Entities:\n"
        f"- {{Tên thực thể}}: {{Loại thực thể}}\n\n"
        f"Văn bản:\n\"{text}\"\n\n"
        f"Kết quả:\n"
        f"Entities:\n"
        f"- {{Tên thực thể}}: {{Loại thực thể}}\n...\n\n"
        f"Hướng dẫn:\n"
        f"- Thực thể là các đối tượng hoặc khái niệm riêng biệt (ví dụ: hàm số, biến, điều kiện, kết quả).\n"
        f"- Đảm bảo các thuật ngữ toán học được diễn giải chính xác (ví dụ: sửa lỗi như 'Delta < 0' ngụ ý '1 nghiệm' thành '0 nghiệm').\n"
        f"- Khi xử lý nhiều tài liệu, hợp nhất các thực thể chỉ cùng một khái niệm (ví dụ: 'Hàm số bậc 2' và 'Hàm số bậc hai' là một thực thể).\n"
        f"- Chuẩn hóa tên thực thể dựa trên ý nghĩa toán học (ví dụ: 'Delta', 'Discriminant', và '\( \Delta \)' là cùng một thực thể).\n"
        f"Ghi chú:\n"
        f"Đây là tài liệu về hàm số bậc 2; nội dung có thể tương tự, nhưng phương pháp giảng dạy và cách trình bày có thể khác nhau.\n\n"
        f"Ví dụ:\n"
        f"Văn bản:\n'Ví dụ về hàm số bậc 2: \( y = ax^2 + bx + c \) (với \( a \neq 0 \)) là hàm số bậc 2. Nếu \( \Delta = b^2 - 4ac > 0 \), hàm số có 2 nghiệm tương ứng. Nếu \( \Delta = b^2 - 4ac = 0 \), hàm số có 1 nghiệm tương ứng. Nếu \( \Delta = b^2 - 4ac < 0 \), hàm số không có nghiệm thực.'\n"
        f"Kết quả:\n"
        f"Entities:\n"
        f"- Hàm số bậc 2: Hàm số\n"
        f"- Delta: Delta\n"
        f"- Tham số a: Hệ số\n"
        f"- Tham số b: Hệ số\n"
        f"- Tham số c: Hệ số\n"
        f"- Công thức tính Delta: Công thức\n"
        f"- Delta > 0: Điều kiện\n"
        f"- Delta = 0: Điều kiện\n"
        f"- Delta < 0: Điều kiện\n"
        f"- Có 2 nghiệm thực: Kết quả\n"
        f"- Có 1 nghiệm thực: Kết quả\n"
        f"- Không có nghiệm thực: Kết quả\n"
    )

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": prompt
                    }
                ]
            },
        ],
        response_format={
            "type": "text"
        },
        temperature=0.5,
        max_tokens=2048,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0,
    )

    # Phân tích kết quả để trả về danh sách entities
    output = response.choices[0].message.content
    entities = []
    lines = output.split('\n')
    for line in lines:
        if line.strip().startswith('- '):
            try:
                name = line.strip()[2:].split(': ')[0]
                entities.append({name.strip()})
            except ValueError:
                continue
    return entities
```

```python
def retriever_similar(query, k):
  retriever = store.as_retriever(
        search_type="similarity_score_threshold",
        search_kwargs={
            "k": k,
            "score_threshold": 0.8,
            "filter": None
        }
    )
  return retriever.invoke(query)
```

```python
def format_context(docs):
    text = ''
    for doc in docs:
        text += doc.page_content + '\n'
    return text
```

```python
def QA(query, llm, k):
  llm_chain = llm
  result_docs = retriever_similar(query, k)
  docs_format = format_context(result_docs)
  response = None
  try:
    response = llm_chain.invoke({ "query": query,
                                 "context": docs_format })
  except Exception as e:
    print("Error:", e)
  return response['text']
```

```python
from neo4j import GraphDatabase

driver = GraphDatabase.driver("bolt://localhost:7687", auth=("neo4j", "password"))

def get_graph_data():
    with driver.session() as session:
        # Get nodes
        nodes_query = "MATCH (n:Entity) RETURN id(n) AS node_id, n.name AS name"
        nodes = session.run(nodes_query)
        node_mapping = {record["node_id"]: record["name"] for record in nodes}

        # Get edges and edge types
        edges_query = "MATCH (n)-[r]->(m) RETURN id(n) AS source, id(m) AS target, type(r) AS relationship_type"
        edges = session.run(edges_query)
        edge_list = [(record["source"], record["target"], record["relationship_type"]) for record in edges]

        # Extract all unique relationship types
        unique_relationship_types = {e[2] for e in edge_list}

        return node_mapping, edge_list, unique_relationship_types

node_mapping, edge_list, unique_relationship_types = get_graph_data()

# Print the unique relationship types
print("Unique Relationship Types:")
for rel_type in unique_relationship_types:
    print(rel_type)
```

```python
import torch
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
import random

# Step 1: Create edge_index
edge_index = torch.tensor([[e[0], e[1]] for e in edge_list], dtype=torch.long).t().contiguous()

# Features for nodes (one-hot encoding)
num_nodes = len(node_mapping)
features = torch.eye(num_nodes)  # One-hot encoding for each node

# Split edges into training and validation sets
num_edges = edge_index.size(1)
indices = list(range(num_edges))
random.shuffle(indices)
split_idx = int(0.8 * num_edges)  # 80% for training, 20% for validation

train_indices = indices[:split_idx]
val_indices = indices[split_idx:]

train_edge_index = edge_index[:, train_indices]
val_edge_index = edge_index[:, val_indices]

# Step 2: Create Data objects for training and validation
train_data = Data(x=features, edge_index=train_edge_index)
val_data = Data(x=features, edge_index=val_edge_index)
```

```python
# Step 3: Define Graph Autoencoder for Nodes Only
class GAE(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, embedding_dim):
        super(GAE, self).__init__()
        self.encoder1 = GCNConv(input_dim, hidden_dim)  # Initializes a GCN layer: (N, N) -> (N, H)
        self.encoder2 = GCNConv(hidden_dim, embedding_dim)  # Initializes a GCN layer: (N, H) -> (N, E)

    def encode(self, x, edge_index):
        # x: (N, N), edge_index: (2, M)
        x = F.relu(self.encoder1(x, edge_index))  # (N, N) -> (N, H)
        x = self.encoder2(x, edge_index)  # (N, H) -> (N, E)
        return x  # Node embeddings: (N, E)

    def decode(self, z, edge_index):
        # z: (N, E), edge_index: (2, M)
        src, tgt = edge_index  # src: (M,), tgt: (M,)
        return (z[src] * z[tgt]).sum(dim=1)  # (M, E) -> (M,)

    def forward(self, x, edge_index):
        # x: (N, F_in), edge_index: (2, M)
        z = self.encode(x, edge_index)  # Node embeddings: (N, E)
        reconstructed = self.decode(z, edge_index)  # Reconstruct edges: (M,)
        return z, reconstructed
```

```python
# Step 4: Initialize the model
input_dim = features.size(1)
hidden_dim = 16
embedding_dim = 8
model = GAE(input_dim, hidden_dim, embedding_dim)

# Step 5: Define optimizer and loss function
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
def loss_function(reconstructed, edge_index):
    # Binary cross-entropy loss for adjacency reconstruction
    # Một tensor toàn giá trị 1 với chiều dài bằng số lượng edges (M) trong đồ thị.
    target = torch.ones(edge_index.size(1))  # All edges exist
    # Sự khác biệt giữa logits dự đoán (reconstructed) và các giá trị thực (target).
    return F.binary_cross_entropy_with_logits(reconstructed, target)

# Step 6: Train the model and validate after each epoch
model.train()
for epoch in range(7000):
    # Training phase
    optimizer.zero_grad()
    embeddings, reconstructed = model(train_data.x, train_data.edge_index)
    train_loss = loss_function(reconstructed, train_data.edge_index)
    train_loss.backward()
    optimizer.step()

    # Validation phase
    model.eval()
    with torch.no_grad():
        val_embeddings, val_reconstructed = model(val_data.x, val_data.edge_index)
        val_loss = loss_function(val_reconstructed, val_data.edge_index)

    # Switch back to train mode for next epoch
    model.train()

    # Print losses
    print(f'Epoch {epoch + 1}, Train Loss: {train_loss.item()}, Validation Loss: {val_loss.item()}')
```
